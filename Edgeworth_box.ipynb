```python
# Define the utility functions manually for plotting
def u1(x, y):
    return x ** 2 + y ** 2

def u2(x, y):
    return x ** 2 + y ** 2
```

```python
# Create a grid for the plot
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)
```

```python
# Define levels for contour plots to represent indifference curves
levels_U = [10, 20, 30, 40]
```

```python
# Setting up the plot
fig, ax = plt.subplots(figsize=(8, 8))

# Plot Consumer 1's indifference curves
contour1 = ax.contour(X, Y, u1(X, Y), levels_U, colors='blue')
ax.clabel(contour1, inline=True, fontsize=10)

# Plot Consumer 2's indifference curves
contour2 = ax.contour(10 - X, 10 - Y, u2(X, Y), levels_U, colors='red', linestyles='dashed')
ax.clabel(contour2, inline=True, fontsize=10)

# Mark initial endowments
plt.scatter(4, 6, color='blue', label='Consumer 1 Endowment')
plt.scatter(6, 4, color='red', label='Consumer 2 Endowment')

# Annotate initial endowments
plt.annotate('Initial Endowment (C1)', (4, 6), textcoords="offset points", xytext=(-10,-10), ha='center')
plt.annotate('Initial Endowment (C2)', (6, 4), textcoords="offset points", xytext=(-10,-10), ha='center')

# Labels and customization
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_xlabel('Good X')
ax.set_ylabel('Good Y')
ax.set_title('Edgeworth Box with Non-Quasi-Concave Utilities')
ax.legend()
plt.grid(True)
plt.show()
```

```python
# Total endowments
total_x = 10
total_y = 10

# Utility functions
def u1(x1, y1):
    return x1 * y1

# Define Consumer 2's function (L-shaped indifference curves)
def u2(x2, y2):
    return np.sqrt(x2) + np.sqrt(y2)

# Create grid
x = np.linspace(0, total_x, 100)
y = np.linspace(0, total_y, 100)
X, Y = np.meshgrid(x, y)

# Contour levels
levels1 = np.array([8, 12, 18, 25])  # Consumer 1
levels2 = np.array([2, 3, 4])        # Consumer 2

# Plot setup
fig, ax = plt.subplots(figsize=(8, 8))

# Plot Consumer 1's indifference curves
iu1 = ax.contour(X, Y, u1(X, Y), levels1, colors='blue')
ax.clabel(iu1, inline=True, fontsize=10)

# Plot Consumer 2's indifference curves, with kinks
ix2 = ax.contour(total_x - X, total_y - Y, u2(X, Y), levels2, colors='red', linestyles='dashed')
ax.clabel(ix2, inline=True, fontsize=10)

# Mark initial endowments
plt.scatter(8, 2, color='blue', label='Consumer 1 Endowment')
plt.scatter(2, 8, color='red', label='Consumer 2 Endowment')

# Annotate initial endowments
plt.annotate('Initial Endowment (C1)', (8, 2), textcoords="offset points", xytext=(-10,-10), ha='center')
plt.annotate('Initial Endowment (C2)', (2, 8), textcoords="offset points", xytext=(-10,-10), ha='center')

# Add labels and legend
ax.set_xlabel('Good X')
ax.set_ylabel('Good Y')
ax.legend()
ax.set_title('Edgeworth Box Example with Correspondence')
plt.xlim(0, total_x)
plt.ylim(0, total_y)
plt.axvline(total_x, color='grey', linestyle='--')
plt.axhline(total_y, color='grey', linestyle='--')
plt.grid(True)
plt.show()
```

```python
from pyEdgeworthBox import EdgeBox as EB
```

```python
u_ann = lambda water, bread: water**0.6 * bread**0.4 # more thirsty than hungry Anna
u_bob = lambda water, bread: water**0.1 * bread**0.9 # more hungry than thirsty Bob
```

```python
IE = [ # initial endowment matrix
    # water (l), bread (kg)
    [4        ,    8],   # Anna
    [11        ,    2]    # Bob
    ]

print("There are %s liter of water, and %s kg of bread in the world" % (IE[0][0] + IE[1][0], IE[0][1] + IE[1][1]))
```

```python
box = EB(
    # what ppl want:
    u1 = u_ann
    , u2 = u_bob
    # what ppl have:
    , IE1 = IE[0]
    , IE2 = IE[1]
    )
```

```python
## options for graphs: ['utility', 'pareto', 'budget', 'core', 'eq']
#box.plot()
#box.plot(graphs = ['pareto', 'core', 'utility'])
#box.plot(graphs = ['pareto', 'core', 'utility'])
box.plot(graphs = ['budget', 'eq'])
```

```python
"Anna Started with %s l of water and %s kg of bread, ended up with %s l of water and %s kg of bread" % (*box.IE1, *box.EQ1)
```

```python
"Bob Started with %s l of water and %s kg of bread, ended up with %s l of water and %s kg of bread" % (*box.IE2, *box.EQ2)
```

```python
box2 = EB(u1 = u_ann
         , u2 = lambda water, bread: water**0.6 * bread**0.4
         , IE1 = IE[0]
         , IE2 = IE[1]
         )
box.plot()
box2.plot()
```
